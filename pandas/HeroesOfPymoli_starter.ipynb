```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
unique_players=purchase_data.groupby(['SN']).count()
total_players = unique_players["Purchase ID"].count()
list(purchase_data)
print(total_players)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#list(purchase_data)
unique_items=purchase_data.groupby(['Item ID']).count()
unique_purchases=purchase_data.groupby(['Purchase ID']).count()
average_price=purchase_data.groupby(['Purchase ID']).mean()


a = unique_items["Purchase ID"].count()
b = unique_purchases["Item ID"].count()
c = average_price["Price"].mean()

#out_data = pd.DataFrame({"Unique Items":1})
out_data = pd.DataFrame({"Unique Items":[a], 
                    "Average Price":[round(c,2)],
                    "Number of Purchases":[b], 
                    "Total Revenue":[c*b],                   
                   }) 

print(out_data)


#unique_purchases["Item ID"].count
#	Number of Unique Items 	Average Price 	Number of Purchases 	Total Revenue
#0 	183 	$3.05 	780 	$2,379.77
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_male = purchase_data.loc[purchase_data["Gender"] == 'Male', :]
purchase_female = purchase_data.loc[purchase_data["Gender"] == 'Female', :]
purchase_other = purchase_data.loc[purchase_data["Gender"] == 'Other / Non-Disclosed', :]

total_sales_male = purchase_male["SN"].count()
total_sales_female = purchase_female["SN"].count()
total_sales_other = purchase_other["SN"].count()

purchase_male_avg = round(purchase_male[["Price"]].mean().mean(),2)
purchase_female_avg = round(purchase_female[["Price"]].mean().mean(),2)
purchase_other_avg = round(purchase_other[["Price"]].mean().mean(),2)

purchase_male_total = round(purchase_male[["Price"]].sum().sum(),2)
purchase_female_total = round(purchase_female[["Price"]].sum().sum(),2)
purchase_other_total = round(purchase_other[["Price"]].sum().sum(),2)

total_different_males_0 = purchase_male.groupby(["SN"]).count()
total_different_males = round(total_different_males_0[["Gender"]].count().sum(),2)



total_different_females_0 = purchase_female.groupby(["SN"]).count()
total_different_females = round(total_different_females_0[["Gender"]].count().sum(),2)

total_different_other_0 = purchase_other.groupby(["SN"]).count()
total_different_other = round(total_different_other_0[["Gender"]].count().sum(),2)

Purchase_by_gender = pd.DataFrame({
                    "Gender":["Male","Female","Other"],
                    "Purchase Count":[round(total_sales_male,2),total_sales_female,total_sales_other], 
                    "Average Purchase Price":[round(purchase_male_avg,2),purchase_female_avg,purchase_other_avg],
                    "Total Purchase Value":[purchase_male_total,purchase_female_total,purchase_other_total], 
                    "Avg Total Purchase per Person":[round(purchase_male_total/total_different_males,2),round(purchase_female_total/total_different_females,2),round(purchase_other_total/total_different_other,2)],                   
                   }) 

Purchase_by_gender.head()
#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Avg Total Purchase per Person

#Gender Total Sales	Average Sale Price	Total Revenue 
#Female 	113 	$3.20 	$361.94 	$4.47
#Male 	652 	$3.02 	$1,967.64 	$4.07
#Other / Non-Disclosed 	15 	$3.35 	$50.19 	$4.56
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
ages_demographics = purchase_data
ages_demographics["Age2"] = purchase_data["Age"]
ages_demographics = ages_demographics.groupby(['SN','Age2']).mean()

#ages_demographics.head()

ages_demographics['age_bins'] = pd.cut(x=ages_demographics['Age'], bins=[0, 9, 14, 19, 24, 29, 34, 39, 100], labels=['<10', '10-14', '15-19','20-24','25-29','30-34','35-39','40+'])
ages_demographics["ctr"] = ages_demographics["Age"]
ages_demographics = ages_demographics[["age_bins","ctr"]].groupby(['age_bins']).count()

ages_demographics["Percentage of Players"] = round((100/total_players)*ages_demographics["ctr"],2)


ages_demographics_save = ages_demographics[['ctr']].copy()


ages_demographics.head(600)

#	Total Count 	Percentage of Players
#<10 	17 	2.95
#10-14 	22 	3.82
#15-19 	107 	18.58
#20-24 	258 	44.79
#25-29 	77 	13.37
#30-34 	52 	9.03
#35-39 	31 	5.38
#40+ 	12 	2.08
```
