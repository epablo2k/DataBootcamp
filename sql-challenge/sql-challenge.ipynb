```python
myFrames = pd.read_sql_query('select title,salary from titles inner join title_categories on title_categories.title_cat_id=titles.title_cat_id inner join salaries on salaries.emp_no=titles.emp_no', connection)
myFrames.head()
```

```python
salarybytitle = myFrames.groupby('title').mean()
```
